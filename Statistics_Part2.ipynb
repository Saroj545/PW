```python
#1. What is hypothesis testing in statistics?
'''
Hypothesis testing is a fundamental statistical method employed in various fields, including data science, machine learning, and statistics, to make informed decisions
based on empirical evidence. It involves formulating assumptions about population parameters using sample statistics and rigorously evaluating these assumptions against
collected data. At its core, hypothesis testing is a systematic approach that allows researchers to assess the validity of a statistical claim about an unknown population parameter.
'''
```

```python
#2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
'''
A null hypothesis is a statement in statistical testing that proposes "no effect" or "no significant difference" between variables, essentially assuming
there is no relationship between them, while the alternative hypothesis is the opposing statement that suggests a real effect or difference does exist between variables,
which is what the researcher is typically trying to prove by rejecting the null hypothesis.

Null hypothesis (H0):
Represents the default assumption, typically stating "no difference" or "no relationship."

Usually denoted by "H0"

Researchers generally aim to reject the null hypothesis to support their research question.

Alternative hypothesis (Ha):
Represents the researcher's prediction or claim about a significant effect or relationship

Usually denoted by "Ha" or "H1"

Accepted if enough evidence is found to reject the null hypothesis

Example:
Null hypothesis:
"There is no difference in test scores between students who studied with a new method and those who studied with the traditional method."

Alternative hypothesis:
"Students who studied with the new method will have significantly higher test scores compared to those who studied with the traditional method
'''
```

```python
#3. What is the significance level in hypothesis testing, and why is it important?
'''
In hypothesis testing, the "significance level" (denoted by the Greek letter alpha, α) represents the probability of incorrectly rejecting a null hypothesis
when it is actually true, essentially setting a threshold for how much evidence is needed to consider a result statistically significant; it is important because
it helps researchers determine whether their findings are likely due to chance or a true effect, allowing them to make informed decisions about rejecting or accepting
their null hypothesis based on a pre-defined level of risk for making a Type I error.

Interpretation:
A significance level of 0.05 means there is a 5% chance of rejecting the null hypothesis when it is true.
Choosing the level:
Researchers typically select a significance level based on the field of study and the potential consequences of making a Type I error.
Relationship with p-value:
When conducting a hypothesis test, the calculated p-value is compared to the chosen significance level. If the p-value is less than the significance level, the null hypothesis is rejected.
'''
```

```python
#4. What does a P-value represent in hypothesis testing?
'''
In hypothesis testing, a p-value is a probability that represents how likely it is that the observed effect occurred by chance. It's used to decide whether to reject the null hypothesis.

A low p-value means that the null hypothesis is unlikely to be true.
A p-value of 0.05 or less is considered statistically significant.
A p-value greater than 0.05 means that no effect was observed.

How to use a p-value?
Use a statistical program to calculate the p-value.
Compare the p-value to a threshold, such as 0.05.
If the p-value is less than the threshold, reject the null hypothesis.

When to use a p-value?
To determine if a sample estimate differs from a hypothesized value
To find out if an effect is due to chance or a specific effect
'''
```

```python
#5.  How do you interpret the P-value in hypothesis testing?
'''
A P-value in hypothesis testing represents the probability of observing data as extreme as, or more extreme than, what was actually observed,
assuming the null hypothesis is true; essentially, it tells you how likely it is to see your results by chance if there is truly no effect present;
a low P-value indicates strong evidence against the null hypothesis, leading to its rejection, while a high P-value suggests the data is more consistent with the null hypothesis being true.

Key points about interpreting P-values:
Smaller P-value, stronger evidence against the null hypothesis:
The lower the P-value, the less likely it is that the observed data occurred by chance alone, making it more likely to reject the null hypothesis.

Comparison to significance level (alpha):
Typically, a P-value less than 0.05 is considered statistically significant, meaning the null hypothesis is rejected.
Not the same as practical significance:
A statistically significant result (low P-value) does not necessarily mean the effect size is large or practically important.

Example:
If a study finds a P-value of 0.02, this means there is only a 2% chance of observing such data if the null hypothesis (no effect) were true, providing strong evidence to reject the null hypothesis.
'''
```

```python
#6. What are Type 1 and Type 2 errors in hypothesis testing?
'''
Type 1 and type 2 errors are mistakes that can occur when testing a hypothesis. They are also known as alpha and beta errors.

Type 1 error
Also known as a false positive
Occurs when a null hypothesis is rejected when it is actually true

For example, an innocent person being convicted

Type 2 error
Also known as a false negative
Occurs when a null hypothesis is not rejected when it should be
For example, a guilty person not being convicted

Type 1 and type 2 errors are the result of forcing quantitative analysis results into a decision
Reducing interpretations to a dichotomy can lead to misinterpreting study results
It's impossible to completely eliminate either type of error
Reducing one type of error can increase the probability of the other

Examples
Roulette wheel: A roulette wheel may be unbiased but land on red 10 times in a row

Virus test: A patient may test positive for a virus when they are not infected

Emergency room: An emergency crew may think a victim is dead when they are actually alive

'''
```

```python
#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
'''
A one-tailed test in hypothesis testing only examines whether a parameter is significantly greater or less than a certain value (i.e., only one direction),
while a two-tailed test examines whether a parameter is significantly different from a certain value in either direction (greater or less than) - essentially
testing for any change, not a specific direction of change.

One-tailed test:
Used when you have a specific prediction about the direction of the effect (e.g., "Treatment A will result in higher scores than Treatment B").
Only considers one "tail" of the distribution (either the upper or lower tail).

Two-tailed test:
Used when you don't have a specific prediction about the direction of the effect (e.g., "Treatment A will have a different effect on scores than Treatment B").
Considers both tails of the distribution (both upper and lower).
'''
```

```python
#8. What is the Z-test, and when is it used in hypothesis testing?
'''
A Z-test is a statistical hypothesis test used to determine if there is a significant difference between a sample mean and a population mean,
or between the means of two groups, when the population standard deviation is known and the sample size is large (typically considered to be 30 or more data points);
it is primarily used when you can assume a normal distribution due to the large sample size.

Purpose:
To evaluate whether a finding or association is statistically significant by comparing a sample mean to a known population mean.

Conditions for use:
The population standard deviation must be known.
The sample size should be large (generally considered 30 or more).

Alternative test for small sample sizes:
If the population standard deviation is unknown or the sample size is small, a T-test is used instead.
How it works:

Calculating a Z-score:
A Z-score represents how many standard deviations a data point is away from the population mean. The Z-test calculates a Z-score based on the sample mean, population mean,
and population standard deviation.
Interpreting the result:
By comparing the calculated Z-score to a critical value from the standard normal distribution, you can determine whether to reject the null hypothesis
(i.e., whether the observed difference is statistically significant).

Example application:
A researcher wants to test if the average height of students in a school is significantly different from the national average height, where the national average is known. If the sample size is large enough, they can use a Z-test to compare the sample mean height to the national average.
'''
```

```python
#9.  How do you calculate the Z-score, and what does it represent in hypothesis testing?
'''
A Z-score is calculated by subtracting the population mean from a data point and then dividing the result by the population standard deviation,
using the formula: z = (x - μ) / σ; it represents how many standard deviations a data point is away from the mean in a normal distribution,
allowing you to compare data points from different distributions in hypothesis testing and determine if a data point is significantly different
from the expected population mean based on its Z-score value relative to a critical value.

Formula: z = (x - μ) / σ
x: Individual data point
μ: Population mean
σ: Population standard deviation

Interpretation:
Positive Z-score: Data point is above the population mean
Negative Z-score: Data point is below the population mean
Z-score close to 0: Data point is close to the population mean

In hypothesis testing:
Comparing Z-score to critical values:
calculate the Z-score of your sample mean and compare it to a critical Z-score based on your chosen significance level (alpha) to determine whether to reject or fail to reject the null hypothesis.

Decision rule:
If the calculated Z-score falls within the critical region (beyond the critical values), you reject the null hypothesis.
'''
```

```python
#10. What is the T-distribution, and when should it be used instead of the normal distribution?
'''
A T-distribution, also known as Student's t-distribution, is a statistical distribution similar to the normal distribution but with "heavier tails,"
meaning it is more likely to produce extreme values, and is used primarily when dealing with small sample sizes or when the population standard deviation is unknown, unlike the normal distribution which assumes the population standard deviation is known; essentially, use the T-distribution when you have limited data and need to estimate the population standard deviation from the sample itself.

Heavier tails:
Compared to the normal distribution, the T-distribution has more probability mass in the tails, which means extreme values are more likely to occur when using a small sample size.

Degrees of freedom:
The shape of the T-distribution is determined by the "degrees of freedom" which is calculated as the sample size minus one (n-1).

When to use it:
When the sample size is small (typically less than 30)
When the population standard deviation is unknown

In summary, use the T-distribution when you have a small sample size and need to estimate the population standard deviation from the sample data,
while the normal distribution is more suitable for large sample sizes where the population standard deviation is known
'''
```

```python
#11. What is the difference between a Z-test and a T-test?
'''
A Z-test is used when the population standard deviation is known and typically requires a large sample size,
while a T-test is used when the population standard deviation is unknown and is more suitable for smaller sample sizes;
essentially, the key difference lies in whether you know the population standard deviation or need to estimate it from
the sample data, which impacts which distribution (normal for Z-test, t-distribution for T-test) is used for the analysis.

Population Standard Deviation:
Z-test: Assumes the population standard deviation is known.
T-test: Assumes the population standard deviation is unknown and must be estimated from the sample.

Sample Size:
Z-test: Typically used for large sample sizes (n > 30).
T-test: More suitable for small sample sizes (n < 30).

Distribution:
Z-test: Relies on the standard normal distribution (Z-distribution).
T-test: Relies on the Student's t-distribution, which adjusts for smaller sample sizes.
'''
```

```python
#12. What is the T-test, and how is it used in hypothesis testing?
'''
A T-test is a statistical test used in hypothesis testing to determine if there is a significant difference between the means of two groups,
or between a sample mean and a known population mean, particularly when the population standard deviation is unknown and the sample size is small;
it essentially measures how much a group mean deviates from another group mean relative to the variability within the groups, allowing you to
assess whether the observed difference is likely due to chance or a real effect.

Purpose:
To compare the averages (means) of two groups to see if they are statistically different from each other.

When to use:
When the population standard deviation is unknown, or when dealing with small sample sizes.

Types of T-tests:
One-sample t-test: Compares the mean of a single group to a known population mean.

Independent samples t-test: Compares the means of two independent groups.

Paired samples t-test: Compares the means of two related groups, like before and after measurements on the same subjects.

Hypothesis testing with T-test:
Null hypothesis (H0): Typically states that there is no significant difference between the groups.
Alternative hypothesis (H1): States that there is a significant difference between the groups.

Calculating a T-statistic:
The T-test calculates a "t-statistic" by dividing the difference between the group means by the pooled standard error.
This t-statistic is then compared to a critical value from the t-distribution to determine statistical significance.

Example: A researcher wants to test if there is a significant difference in test scores between students who received a
new tutoring program and those who did not. They would use an independent samples t-test to compare the average test scores of the two groups.
'''
```

```python
#13.  What is the relationship between Z-test and T-test in hypothesis testing?
'''
A Z-test and a T-test are both statistical tests used in hypothesis testing to compare means, but the key difference
is that a Z-test is used when the population standard deviation is known and the sample size is large, while a T-test is
used when the population standard deviation is unknown or when the sample size is small; essentially, a Z-test relies
on the normal distribution while a T-test uses the t-distribution, which is more suitable for smaller sample sizes.

Sample size:
Z-test is typically used for large sample sizes (n > 30), while a T-test is used for small sample sizes (n < 30).

Population standard deviation:
Z-test assumes the population standard deviation is known, whereas a T-test estimates the population standard deviation from the sample.

Distribution:
Z-test uses the standard normal distribution to calculate the test statistic, while a T-test uses the t-distribution.
'''
```

```python
#14. What is a confidence interval, and how is it used to interpret statistical results?
'''
A confidence interval (CI) is a range of values that is likely to contain the true value of a population parameter (like a mean or proportion) based on a
sample, providing a measure of uncertainty around a statistical estimate, typically expressed as a percentage like 95% confidence level, meaning that if
the study was repeated multiple times, the true population value would fall within that range in 95% of cases; it is used to interpret statistical results
by indicating the degree of reliability around a point estimate, allowing researchers to understand the potential variability in their findings and assess the significance of the observed effect.

Interpretation:
A 95% confidence interval means that if the study was repeated many times, the true population parameter would fall within the calculated range in 95% of those repetitions.

Components:
A confidence interval is composed of a lower bound and an upper bound, representing the range of values within which the true population parameter is likely to lie.

Confidence level:
The level of confidence (e.g., 95%, 99%) determines how wide the interval is, with higher confidence levels resulting in a wider interval.

Sample size:
Larger sample sizes generally produce narrower confidence intervals, indicating greater precision in the estimate.

How to use confidence intervals in interpretation:
Assessing statistical significance:
If a confidence interval does not include zero (for a difference between groups), it suggests a statistically significant result at the chosen confidence level.

Comparing effect sizes:
The width of the confidence interval can be used to compare the relative precision of different estimates.

Understanding limitations:
Confidence intervals do not tell you the probability that a specific value within the range is the true population value, only the probability that the true value falls within the entire range.
'''
```

```python
#15.  What is the margin of error, and how does it affect the confidence interval?
'''
A "margin of error" is a statistical measure that represents the degree of uncertainty around a survey result or estimate, essentially
indicating how much the true population value could differ from the sample result, while a "confidence interval" is the range of values
within which the true population value is likely to fall, with the margin of error defining the "plus or minus" range around the estimated
value within that confidence interval; in simpler terms, the margin of error is essentially half the width of the confidence interval, meaning
a larger margin of error results in a wider confidence interval, indicating less precision in the estimate.

Interpretation:

When a survey reports a result with a margin of error, it means that the true population value is likely to fall within a certain range around the reported value,
with the margin of error defining the boundaries of that range.

Relationship:
The confidence interval is calculated by taking the point estimate (the sample result) and adding and subtracting the margin of error.

Factors affecting margin of error:
A larger sample size generally leads to a smaller margin of error, meaning greater precision in the estimate.

Confidence level:
The confidence level (e.g., 95%) determines how wide the confidence interval needs to be to capture the true population value with a certain level of certainty.

Example: If a poll reports that 50% of people support a policy with a margin of error of 3%, it means that the true population value is likely between 47% and 53%
'''
```

```python
#16. How is Bayes' Theorem used in statistics, and what is its significance?
'''
Bayes' Theorem is a statistical formula used to calculate the conditional probability of an event, meaning the likelihood of an event occurring given that
another related event has already happened; it essentially allows you to update your beliefs about a situation based on new information, making it a
powerful tool for revising probabilities as new evidence becomes available.

Calculating conditional probability:
The core function of Bayes' Theorem is to calculate the probability of event A occurring given that event B has already occurred, represented as P(A|B).

Updating beliefs with new data:
A significant aspect of Bayes' Theorem is its ability to incorporate prior knowledge (initial beliefs) and update them based on new evidence, leading to a
"posterior probability" that reflects the revised understanding after considering new information.

Applications in various fields:
Due to its capability to handle uncertainty and incorporate new data, Bayes' Theorem is widely used in fields like medicine (diagnosing diseases), machine learning (classification tasks),
finance (risk assessment), and scientific research.

Example application:
Medical diagnosis: If a patient shows symptoms of a disease, Bayes' Theorem can be used to calculate the probability of the patient actually having the disease based on the test results, considering the disease prevalence in the population and the accuracy of the diagnostic test.

Important terms related to Bayes' Theorem:
Prior probability: The initial belief about the probability of an event before new evidence is considered.

Likelihood: The probability of observing the data given a specific hypothesis.

Posterior probability: The updated probability of an event after considering new evidence.
'''
```

```python
#17. What is the Chi-square distribution, and when is it used?
'''
A Chi-square distribution is a statistical distribution used primarily for hypothesis testing, particularly when analyzing categorical data,
to determine if there is a significant difference between observed frequencies and expected frequencies in a dataset, often used in tests
like the "chi-square goodness of fit test" and the "chi-square test of independence" to assess relationships between categorical variables;
its shape is determined by the "degrees of freedom" parameter which is based on the number of categories in the data involved.

Purpose:
Primarily used for hypothesis testing, specifically to compare observed frequencies in categorical data to expected frequencies based on a null hypothesis.

Common tests:
Chi-square goodness of fit test: Checks if a set of observed data fits a theoretical distribution.

Chi-square test of independence: Tests whether two categorical variables are independent of each other.

Key feature - Degrees of Freedom:
The shape of the chi-square distribution is determined by the degrees of freedom, which is a value calculated based on the number of categories in the data.

When to use a Chi-square test:
When analyzing categorical data (e.g., gender, color preference, opinion categories).
When comparing observed frequencies in a dataset to expected frequencies based on a specific hypothesis.
To test for relationships between two categorical variables.
'''
```

```python
#18. What is the Chi-square goodness of fit test, and how is it applied?
'''
A Chi-square goodness of fit test is a statistical test used to determine whether a set of observed data from a sample closely resembles a particular theoretical distribution
(like a normal distribution or binomial distribution), essentially checking if the "fit" between the observed data and the expected distribution is good enough to be considered
statistically significant; it does this by comparing the frequencies of observed categories in the data to the expected frequencies based on the theoretical distribution,
calculating a Chi-square statistic to assess the discrepancy between the two sets of frequencies.

How to apply the Chi-square goodness of fit test:
Define your null hypothesis:
This typically states that the observed data follows the specified theoretical distribution.
Categorize your data:

Divide the data into distinct categories or bins, based on the theoretical distribution.

Calculate expected frequencies:
For each category, calculate the expected number of observations based on the theoretical distribution.

Calculate the Chi-square statistic:
For each category, subtract the observed frequency from the expected frequency, square the difference, and then divide by the expected frequency.
Sum up these values across all categories to get the Chi-square statistic.

Determine the critical value:
Using the degrees of freedom (number of categories minus 1), look up the critical Chi-square value from a Chi-square distribution table at a chosen significance level (e.g., 0.05).

Compare the calculated Chi-square statistic to the critical value:
If the calculated Chi-square statistic is greater than the critical value, reject the null hypothesis, meaning the data does not fit the theoretical distribution well.

Key points about the Chi-square goodness of fit test:
Categorical data: This test is only applicable to categorical data, where observations are classified into distinct categories.
Expected frequencies: It is crucial to have sufficient expected frequencies in each category for the test to be reliable.
Assumptions: The observations in the data should be independent of each other.

Example application:
Testing for a fair dice: You could roll a dice many times and compare the observed frequencies of each number to the expected frequencies (1/6 for each number) to see
if the dice is fair using a Chi-square goodness of fit test.
'''
```

```python
#19. What is the F-distribution, and when is it used in hypothesis testing?
'''
The F-distribution is a continuous probability distribution used in statistics primarily for hypothesis testing when comparing the variances
of two populations, most commonly within the context of an Analysis of Variance (ANOVA) test; it is used to determine if there is a significant difference between the variances of two sample groups, essentially asking if the variability between groups is larger than the variability within groups.

Purpose:
To compare variances between two groups, allowing you to test if there is a significant difference in how spread out the data is between them.

How it's calculated:
The F-statistic is calculated by dividing the variance of one group by the variance of another group, and this ratio is then compared to the F-distribution to determine statistical significance.

Assumptions:
To use the F-distribution, the data from each group is typically assumed to be normally distributed and independent.

Key feature:
The F-distribution is positively skewed and depends on two sets of degrees of freedom, one for each group being compared.
When to use the F-distribution in hypothesis testing:

Comparing variances of two groups:
When you want to test if the variability in one population is significantly different from the variability in another population.

ANOVA tests:
The F-distribution is the primary statistical tool used in ANOVA, which allows you to compare the means of multiple groups simultaneously.
'''
```

```python
#20. What is an ANOVA test, and what are its assumptions?
'''
An ANOVA (Analysis of Variance) test is a statistical method used to compare the means of three or more groups, determining if there are
significant differences between them by analyzing the variance within and between the groups; its primary assumptions are that the data is
normally distributed within each group, samples are randomly selected and independent, and all groups have roughly equal variances (homogeneity of variance).

Purpose:
To assess whether observed differences between group means are statistically significant, meaning they are unlikely to be due to random chance.

Key assumptions:
Normality: The populations from which samples are drawn follow a normal distribution.
Independence: Each observation within a group is independent of other observations.
Homogeneity of variance: The variance within each group is approximately equal.

How it works:
ANOVA calculates a test statistic (F-statistic) by comparing the variance between groups to the variance within groups.
'''
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import math
from scipy.stats import norm

# Function to perform Z-test
def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    # Calculate the standard error
    standard_error = population_std / math.sqrt(sample_size)

    # Calculate the Z score
    z_score = (sample_mean - population_mean) / standard_error

    # Calculate the p-value from the Z score
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

    # Output results
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation based on p-value
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference.")

# Example input
sample_mean = 105       # Example sample mean
population_mean = 100   # Known population mean
population_std = 15     # Known population standard deviation
sample_size = 50        # Sample size

# Perform the Z-test
z_test(sample_mean, population_mean, population_std, sample_size)
```

```python
#2.  Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python
import numpy as np
import math
from scipy.stats import norm

# Function to perform Z-test
def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    # Calculate the standard error
    standard_error = population_std / math.sqrt(sample_size)

    # Calculate the Z score
    z_score = (sample_mean - population_mean) / standard_error

    # Calculate the p-value from the Z score (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Output results
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation based on p-value
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference.")

# Simulate random data (e.g., 100 sample points)
np.random.seed(42)  # For reproducibility
population_mean = 100
population_std = 15
sample_size = 100

# Generate random sample data from a normal distribution
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Calculate the sample mean
sample_mean = np.mean(sample_data)

# Perform the Z-test
print("Performing hypothesis test for sample data...")
z_test(sample_mean, population_mean, population_std, sample_size)
```

```python
#3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean
import math
from scipy.stats import norm
import numpy as np

# Function to perform a one-sample Z-test
def z_test(sample_data, population_mean, population_std, alpha=0.05):
    # Calculate sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate the standard error
    standard_error = population_std / math.sqrt(sample_size)

    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / standard_error

    # Calculate the p-value from the Z score (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

    # Output results
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation based on p-value
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference.")


np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=105, scale=15, size=50)  # Sample from normal distribution

# Known population parameters
population_mean = 100
population_std = 15

# Perform the Z-test
print("Performing one-sample Z-test...")
z_test(sample_data, population_mean, population_std)
```

```python
#5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
def visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05):
    """
    Visualizes Type 1 and Type 2 errors in hypothesis testing.

    Parameters:
    mu0 : float - Population mean under H0
    mu1 : float - Population mean under H1 (alternative hypothesis)
    sigma : float - Population standard deviation
    n : int - Sample size
    alpha : float - Significance level (default is 0.05)
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value (two-tailed test)
    z_critical = stats.norm.ppf(1 - alpha / 2)
    critical_value_low = mu0 - z_critical * se
    critical_value_high = mu0 + z_critical * se

    # X-axis values
    x = np.linspace(mu0 - 4 * se, mu1 + 4 * se, 1000)
    y_h0 = stats.norm.pdf(x, mu0, se)  # Null distribution
    y_h1 = stats.norm.pdf(x, mu1, se)  # Alternative distribution

    plt.figure(figsize=(10, 6))

    # Plot H0 and H1 distributions
    plt.plot(x, y_h0, label="Null Distribution ($H_0$)", color='black')
    plt.plot(x, y_h1, label="Alternative Distribution ($H_A$)", color='blue')

    # Shade Type 1 Error (Alpha Region)
    plt.fill_between(x, y_h0, where=(x <= critical_value_low) | (x >= critical_value_high),
                     color='red', alpha=0.3, label="Type 1 Error (α)")

    # Shade Type 2 Error (Beta Region)
    plt.fill_between(x, y_h1, where=(x > critical_value_low) & (x < critical_value_high),
                     color='blue', alpha=0.3, label="Type 2 Error (β)")

    # Critical values
    plt.axvline(critical_value_low, color='black', linestyle='dotted', label=f"Critical Values: {critical_value_low:.2f}, {critical_value_high:.2f}")
    plt.axvline(critical_value_high, color='black', linestyle='dotted')

    # Labels and title
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

    # Compute Beta (Type 2 error probability)
    beta = stats.norm.cdf(critical_value_high, mu1, se) - stats.norm.cdf(critical_value_low, mu1, se)

    return {"Type 1 Error (α)": alpha, "Type 2 Error (β)": beta}

# Example usage with parameters:
mu0 = 100  # Null hypothesis mean
mu1 = 105  # Alternative hypothesis mean
sigma = 15  # Population standard deviation
n = 30  # Sample size

visualize_type1_type2_errors(mu0, mu1, sigma, n)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results
from scipy.stats import ttest_ind

def perform_t_test(sample1, sample2, alpha=0.05):
    """
    Performs an independent two-sample T-test and interprets the results.

    Parameters:
    sample1 : list or array - Data for group 1
    sample2 : list or array - Data for group 2
    alpha : float - Significance level (default is 0.05)

    Returns:
    Dictionary containing test results.
    """
    # Perform independent T-test
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)  # Welch's T-test (unequal variances)

    # Interpretation
    if p_value < alpha:
        result = "Reject the null hypothesis: The means are significantly different."
    else:
        result = "Fail to reject the null hypothesis: No significant difference in means."

    return {
        "T-statistic": t_stat,
        "P-value": p_value,
        "Decision": result
    }

# Example data (two independent groups)
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=100, scale=15, size=30)  # Mean=100, Std=15, Size=30
group2 = np.random.normal(loc=110, scale=15, size=30)  # Mean=110, Std=15, Size=30

# Perform T-test
t_test_results = perform_t_test(group1, group2)
t_test_results
```

```python
#7.  Perform a paired sample T-test using Python and visualize the comparison results
from scipy.stats import ttest_rel
import seaborn as sns

def perform_paired_t_test(before, after, alpha=0.05):
    """
    Performs a paired sample T-test and visualizes the comparison.

    Parameters:
    before : list or array - Data before treatment/intervention
    after : list or array - Data after treatment/intervention
    alpha : float - Significance level (default is 0.05)

    Returns:
    Dictionary containing test results.
    """
    # Perform paired T-test
    t_stat, p_value = ttest_rel(before, after)

    # Interpretation
    if p_value < alpha:
        result = "Reject the null hypothesis: Significant difference between before and after."
    else:
        result = "Fail to reject the null hypothesis: No significant difference."

    # Visualization of before vs after using Seaborn
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=[before, after], palette=["blue", "green"])
    plt.xticks([0, 1], ["Before", "After"])
    plt.ylabel("Values")
    plt.title("Paired Sample Comparison")
    plt.show()

    return {
        "T-statistic": t_stat,
        "P-value": p_value,
        "Decision": result
    }

# Example Data: Before and After Treatment Scores
np.random.seed(42)  # For reproducibility
before_treatment = np.random.normal(loc=100, scale=10, size=30)
after_treatment = before_treatment + np.random.normal(loc=5, scale=5, size=30)  # Adding some effect

# Perform Paired T-test
paired_t_test_results = perform_paired_t_test(before_treatment, after_treatment)
paired_t_test_results
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
def confidence_interval(sample, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Parameters:
    sample : list or array - Sample data
    confidence : float - Confidence level (default is 95%)

    Returns:
    Tuple containing (lower bound, upper bound) of the confidence interval.
    """
    n = len(sample)  # Sample size
    sample_mean = np.mean(sample)  # Sample mean
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)

    # Get the critical t-value for the given confidence level
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample
ci_result = confidence_interval(sample_data)
ci_result
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data
def margin_of_error(sample, confidence=0.95):
    """
    Calculates the margin of error for a given confidence level using sample data.

    Parameters:
    sample : list or array - Sample data
    confidence : float - Confidence level (default is 95%)

    Returns:
    Margin of error (MoE)
    """
    n = len(sample)  # Sample size
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)

    # Get the critical t-value for the given confidence level
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Calculate margin of error
    moe = t_critical * (sample_std / np.sqrt(n))

    return moe

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample data

# Compute Margin of Error
moe_result = margin_of_error(sample_data)
moe_result


```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior, likelihood, evidence):
    """
    Computes the posterior probability using Bayes' Theorem.

    Parameters:
    prior : float - Prior probability P(H)
    likelihood : float - Likelihood P(E|H)
    evidence : float - Total probability P(E)

    Returns:
    Posterior probability P(H|E)
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Example Scenario: Medical Diagnosis
# Suppose we are testing for a rare disease.
prior_disease = 0.01  # P(Disease) = 1%
likelihood_positive_test = 0.95  # P(Positive Test | Disease) = 95%
false_positive_rate = 0.05  # P(Positive Test | No Disease) = 5%
prob_no_disease = 1 - prior_disease  # P(No Disease)

# Total probability of a positive test P(E)
evidence_positive_test = (likelihood_positive_test * prior_disease) + (false_positive_rate * prob_no_disease)

# Compute posterior probability P(Disease | Positive Test)
posterior_disease_given_positive = bayes_theorem(prior_disease, likelihood_positive_test, evidence_positive_test)
posterior_disease_given_positive
```

```python
#16. Implement an F-test using Python to compare the variances of two random samples
def f_test(sample1, sample2):
    """
    Performs an F-test to compare variances of two independent samples.

    Parameters:
    sample1 : array - First sample data
    sample2 : array - Second sample data

    Returns:
    F-statistic and p-value
    """
    var1 = np.var(sample1, ddof=1)  # Sample variance (unbiased)
    var2 = np.var(sample2, ddof=1)

    # Compute F-statistic (larger variance / smaller variance)
    if var1 > var2:
        f_stat = var1 / var2
    else:
        f_stat = var2 / var1

    # Compute p-value using F-distribution
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1
    p_value = 2 * (1 - stats.f.cdf(f_stat, df1, df2))  # Two-tailed test

    return f_stat, p_value

# Generate two random samples
np.random.seed(42)
sample_A = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
sample_B = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std=15

# Perform F-test
f_statistic, p_val = f_test(sample_A, sample_B)

# Results
f_test_results = {
    "F-statistic": f_statistic,
    "P-value": p_val,
    "Conclusion": "Reject H0 (Variances are different)" if p_val < 0.05 else "Fail to reject H0 (Variances are equal)"
}

f_test_results
```

```python
#17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results
def anova_test(*groups):
    """
    Performs a one-way ANOVA test on multiple groups.

    Parameters:
    *groups : arrays - Sample data for each group

    Returns:
    F-statistic and p-value
    """
    f_stat, p_value = stats.f_oneway(*groups)
    return f_stat, p_value

# Generate three random samples (populations with different means)
np.random.seed(42)
group_1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
group_2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10
group_3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10

# Perform ANOVA test
f_statistic, p_val = anova_test(group_1, group_2, group_3)

# Results
anova_results = {
    "F-statistic": f_statistic,
    "P-value": p_val,
    "Conclusion": "Reject H0 (At least one mean is different)" if p_val < 0.05 else "Fail to reject H0 (All means are equal)"
}

anova_results
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
def check_anova_assumptions(*groups):
    """
    Checks the assumptions of ANOVA: Normality and Homogeneity of Variances.

    Parameters:
    *groups : arrays - Sample data for each group

    Returns:
    Dictionary with p-values for normality and variance tests.
    """
    results = {}

    # Check Normality using Shapiro-Wilk Test
    normality_pvalues = [stats.shapiro(group)[1] for group in groups]
    results["Normality P-values"] = normality_pvalues
    results["Normality Conclusion"] = [
        "Pass (Normal)" if p > 0.05 else "Fail (Not Normal)" for p in normality_pvalues
    ]

    # Check Homogeneity of Variances using Levene's Test
    levene_pvalue = stats.levene(*groups)[1]
    results["Levene’s Test P-value"] = levene_pvalue
    results["Variance Conclusion"] = "Pass (Equal Variances)" if levene_pvalue > 0.05 else "Fail (Unequal Variances)"

    return results

# Check assumptions for our ANOVA data (group_1, group_2, group_3)
anova_assumption_results = check_anova_assumptions(group_1, group_2, group_3)
anova_assumption_results
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
# Hypothesis test for population variance using Chi-square distribution

# Known population variance (assumed for testing)
pop_variance = 100
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
# Sample variance
sample_variance = np.var(sample1, ddof=1)  # Unbiased sample variance

# Degrees of freedom
df = len(sample1) - 1

# Chi-square test statistic
chi_square_stat = (df * sample_variance) / pop_variance

# Critical values for a two-tailed test (95% confidence level)
alpha = 0.05  # Significance level
chi_crit_low = stats.chi2.ppf(alpha / 2, df)  # Lower critical value
chi_crit_high = stats.chi2.ppf(1 - alpha / 2, df)  # Upper critical value

# P-value
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

# Conclusion
variance_test_results = {
    "Chi-square Statistic": chi_square_stat,
    "P-value": p_value,
    "Critical Range": (chi_crit_low, chi_crit_high),
    "Conclusion": "Reject H0 (Variance is significantly different)" if p_value < 0.05 else "Fail to reject H0 (Variance is not significantly different)"
}

variance_test_results
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
from statsmodels.stats.proportion import proportions_ztest

def z_test_proportions(success_a, size_a, success_b, size_b, alternative='two-sided'):
    """
    Performs a Z-test for comparing two proportions.

    Parameters:
    success_a (int): Number of successes in group A
    size_a (int): Total sample size of group A
    success_b (int): Number of successes in group B
    size_b (int): Total sample size of group B
    alternative (str): 'two-sided', 'larger', or 'smaller'

    Returns:
    Dictionary with Z-statistic, p-value, and conclusion.
    """
    count = np.array([success_a, success_b])
    nobs = np.array([size_a, size_b])

    # Perform the Z-test for proportions
    z_stat, p_value = proportions_ztest(count, nobs, alternative=alternative)

    # Conclusion
    conclusion = "Reject H0 (Proportions are significantly different)" if p_value < 0.05 else "Fail to reject H0 (No significant difference)"

    return {
        "Z-statistic": z_stat,
        "P-value": p_value,
        "Conclusion": conclusion
    }

# Example usage (simulated data)
results_ztest = z_test_proportions(success_a=50, size_a=200, success_b=65, size_b=220)
results_ztest
```

```python
#26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
# Function to perform F-test for variance comparison
def f_test_variance(sample1, sample2):
    """
    Performs an F-test to compare the variances of two datasets.

    Parameters:
    sample1 (array): First dataset
    sample2 (array): Second dataset

    Returns:
    Dictionary with F-statistic, p-value, and conclusion.
    """
    var1, var2 = np.var(sample1, ddof=1), np.var(sample2, ddof=1)
    df1, df2 = len(sample1) - 1, len(sample2) - 1

    # Compute F-statistic (larger variance / smaller variance)
    f_stat = var1 / var2 if var1 > var2 else var2 / var1
    p_value = 2 * (1 - stats.f.cdf(f_stat, df1, df2))

    # Conclusion
    conclusion = "Reject H0 (Variances are significantly different)" if p_value < 0.05 else "Fail to reject H0 (No significant difference in variances)"

    return {"F-statistic": f_stat, "P-value": p_value, "Conclusion": conclusion}

# Simulated datasets with different variances
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=5, size=30)  # Std Dev = 5
data2 = np.random.normal(loc=50, scale=10, size=30) # Std Dev = 10

# Perform the F-test
f_test_results = f_test_variance(data1, data2)

# Visualization: Boxplot to compare variances
plt.figure(figsize=(8, 6))
sns.boxplot(data=[data1, data2], palette="Set2")
plt.xticks([0, 1], ["Sample 1", "Sample 2"])
plt.title(f"F-Test for Variance Comparison (F={f_test_results['F-statistic']:.2f}, p={f_test_results['P-value']:.5f})")
plt.ylabel("Values")
plt.grid()

# Show plot and results
plt.show()
f_test_results

```
