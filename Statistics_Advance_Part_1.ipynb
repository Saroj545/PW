```python
#1. What is a random variable in probability theory?
'''
A random variable is a function that maps the outcomes of a random experiment to real numbers. It's a way to abstractly represent the outcomes of a random experiment.
Random variables are used to quantify the outcomes of random events.  They are represented by capital italicized letters, like X.  The value of a random variable depends on the outcome of the experiment.
Each random variable has a probability distribution function, which shows the probability of each possible outcome.
Random variables can be discrete or continuous.

Examples
Tossing a coin If X is the number of heads after two coin tosses, then X(HH)=2, X(HT)=1, X(TH)=1, and X(TT)=0.

Rolling dice If X is the sum of three dice rolls, then X could be 3, 18, or any value between them.

Burning a light bulb If Y is the lifetime of a light bulb in hours, then Y can be any positive real number between 0 and 100.
'''
```

```python
#2. What are the types of random variables?
'''
The three main types of random variables are discrete, continuous, and mixed.

Discrete random variables

-Have a finite number of possible values
-The sample space is countable
-Each outcome has a specific probability
-The total probability of all outcomes is 1
An example is rolling a six-sided dice

Continuous random variables

-Have an infinite number of possible values
-The sample space is continuous
-The probability of any single value is zero
-The variable is defined over an interval of values
Examples include height, weight, and time

Mixed random variables
A combination of discrete and continuous random variables

A random variable is a function that assigns a numerical value to each outcome of an experiment. The value of a random variable varies with each trial of the experiment.
'''
```

```python
#3. What is the difference between discrete and continuous distributions?
'''
A discrete distribution refers to a probability distribution where a variable can only take on a finite number of distinct values (usually countable),
while a continuous distribution allows a variable to take on any value within a specified range, meaning there are infinitely many possible outcomes
within that range; essentially, discrete data is counted, while continuous data is measured.

discrete distributions:
-Countable outcomes: You can count the number of possible values a discrete variable can take.
-Examples: Number of heads in 5 coin flips, number of cars passing a certain point on a road in an hour.
-Probability mass function: Used to calculate the probability of each individual value.

continuous distributions:
-Infinitely many possible values: A continuous variable can take on any value within a given range, meaning there are infinitely many possibilities.
-Examples: Height of a person, weight of an object, time taken to complete a task.
-Probability density function: Used to calculate the probability of a variable falling within a certain range.
'''
```

```python
#5.  How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
'''
A cumulative distribution function (CDF) represents the probability that a random variable will be less than or equal to a specific value,
while a probability distribution function (PDF) represents the probability density at a specific value, meaning it shows the likelihood of a
random variable taking on that exact value; essentially, a CDF accumulates probabilities up to a certain point, whereas a PDF gives the probability at a single point.

Key differences:
Information provided:
A CDF gives the cumulative probability up to a certain point, while a PDF gives the probability density at a single point.

Interpretation:
A CDF tells you the chance of observing a value less than or equal to a given value, while a PDF tells you the likelihood of observing a value exactly equal to a given value.

Mathematical relationship:
Integration: To calculate a CDF, you integrate the PDF over a range from negative infinity to the desired value.
Differentiation: To calculate a PDF, you differentiate the CDF.

Example:
Imagine a distribution of heights in a population.
PDF:
The PDF at a height of 1.8 meters would represent the probability of someone in the population being exactly 1.8 meters tall.
CDF:
The CDF at a height of 1.8 meters would represent the probability of someone in the population being 1.8 meters tall or shorter
'''
```

```python
#8. What is a binomial distribution and how is it used in probability?
'''
A binomial distribution is a probability distribution used to calculate the likelihood of getting a specific number of "successes" in a
fixed number of independent trials, where each trial has only two possible outcomes (like "success" or "failure") and the probability of success remains constant throughout all trials; it's commonly used in situations where you want to find the probability of a certain event occurring a specific number of times within a set number of attempts, like flipping a coin multiple times and counting how many heads you get.

Key points about binomial distribution:
Two possible outcomes: Each trial in a binomial distribution can only result in one of two outcomes, typically called "success" and "failure".
Fixed number of trials: You need to specify the total number of independent trials (n) that are being conducted.
Constant probability of success: The probability of success (p) must be the same for each trial.

How it's used in probability:
Calculating probabilities: The binomial distribution formula allows you to calculate the probability of getting exactly "k" successes in "n" trials, given the probability of success on each trial (p).
Example:
Flipping a coin: If you flip a fair coin 10 times, what is the probability of getting exactly 5 heads?
Here, "n" (number of trials) is 10, "k" (desired number of successes) is 5, and "p" (probability of success on each trial) is 0.5.

Applications of binomial distribution:
Quality control: Checking the proportion of defective items in a batch of products
Genetics: Calculating the probability of inheriting a specific trait based on parental genotypes
Polling: Estimating the likelihood of a candidate winning an election based on survey results
Medical research: Analyzing the success rate of a new treatment in a clinical trial
'''
```

```python
#13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
'''
The Central Limit Theorem (CLT) states that as you take larger and larger random samples from a population, the distribution of the sample means will tend to approximate a
normal distribution (bell curve), regardless of the original population distribution shape, making it a crucial tool in statistics for making inferences about a population
based on sample data; essentially, it allows us to use the well-known properties of the normal distribution to analyze data even when the original population distribution
is not normal, as long as the sample size is sufficiently large.

Application: It enables statisticians to use the normal distribution to calculate confidence intervals and perform hypothesis tests on population means, even when the
original population distribution is unknown or not normally distributed.
Sample size matters: The CLT becomes more accurate as the sample size increases.
Importance in statistical inference: Because of its ability to simplify analysis by approximating complex distributions with the normal distribution, the CLT is considered a
fundamental principle in statistical inference.
'''
```

```python
#14.  How does the Central Limit Theorem relate to the normal distribution?
'''
The Central Limit Theorem (CLT) states that as the sample size from a population increases, the distribution of the sample means will tend to approximate a normal distribution,
regardless of the original population distribution shape, meaning that even if the original data isn't normally distributed, the average of many samples taken from that population
will be approximately normally distributed when the sample size is large enough; essentially, the CLT connects the concept of sampling to the normal distribution, allowing us to
use normal distribution properties to analyze data from various populations with sufficiently large sample sizes.

relationship between CLT and normal distribution:
Approximation:
The CLT tells us that the distribution of sample means will approach a normal distribution as the sample size grows, not that the original population data needs to be normally distributed.
Importance in statistics:
This property is crucial in statistics because it allows us to use statistical methods based on the normal distribution to analyze data from various populations, even when the original
data is not normally distributed, as long as the sample size is large enough.
Conditions for CLT:
For the CLT to hold, the samples must be independently drawn from the population and the population variance must be finite.
'''
```

```python
#15. What is the application of Z statistics in hypothesis testing?
'''
A Z-statistic is used in hypothesis testing to determine whether a finding or association between variables is statistically significant, particularly when
comparing the mean of a sample to a known population mean, provided the population standard deviation is known or the sample size is large enough; essentially,
it helps decide whether to reject a null hypothesis based on the calculated Z-score compared to a critical value.

Z-statistics in hypothesis testing:

When to use:
A Z-test is appropriate when the population standard deviation is known and the sample size is large (generally considered to be 30 or more data points).
Comparison to population mean:
The Z-statistic essentially calculates how many standard deviations away a sample mean is from the population mean.

Interpreting the Z-score:
A large absolute value of the Z-score indicates a significant difference between the sample mean and the population mean, leading to a rejection of the null hypothesis.

Limitations:
If the population standard deviation is unknown or the sample size is small, a T-test is more suitable.

Example application:
A researcher wants to test if the average height of adults in a city is significantly different from the national average of 5'8". They collect a large
sample of heights and use a Z-test to compare the sample mean to the national average, determining whether the difference is statistically significant.
'''
```

```python
#17. What are point estimates and interval estimates in statistics?
'''
a point estimate is a single value used to estimate a population parameter based on sample data, while an interval estimate provides a range
of values within which the population parameter is likely to fall, offering a measure of uncertainty around the point estimate; the most common type of interval estimate is a confidence interval.
Key points about point and interval estimates:

Point estimate:
Provides a single "best guess" for the population parameter based on the sample data.
Example: If you calculate the average height of a sample of people, that average is a point estimate for the average height of the whole population.

Interval estimate:
Provides a range of values where the true population parameter is likely to lie, considering a certain level of confidence.
Example: Saying that the average height of the population is likely between 5.5 and 6 feet with a 95% confidence level.
'''
```

```python
#18. What is the significance of confidence intervals in statistical analysis?
'''
A confidence interval in statistical analysis is significant because it provides a range of values within which the true population parameter is likely to fall,
offering a measure of uncertainty around a sample statistic, and indicating how reliable the estimate is based on the data, unlike just a single point estimate;
a narrower confidence interval signifies greater precision in the estimate while a wider interval indicates more uncertainty.
confidence intervals:

Interpretation:
A 95% confidence interval means that if the study was repeated multiple times, the true population parameter would fall within the calculated range in 95% of those repetitions.
Provides context to statistical significance:
Confidence intervals can be used to assess the practical significance of a result even if it is statistically significant, as they show the magnitude of the effect.

Factors affecting width:
The width of a confidence interval is influenced by the sample size - larger sample sizes generally lead to narrower confidence intervals, signifying greater precision.

Comparison with p-values:
While p-values indicate whether a result is statistically significant, confidence intervals provide information about the likely range of the true population parameter.
'''
```

```python
#22. What is the concept of expected value in a probability distribution?
'''
In a probability distribution, the "expected value" refers to the weighted average of all possible outcomes of a random variable, where each
outcome is weighted by its corresponding probability; essentially, it represents the "average" value you would anticipate seeing if you repeated
the experiment a large number of times, often also called the "mean" of the distribution.
Key points about expected value:

Calculation:
To calculate the expected value, you multiply each possible outcome by its probability and then sum up all those products.
Formula:
For a discrete random variable X, the expected value is denoted as E(X) = ∑(x * P(x)), where x represents the possible values and P(x) is the probability of each value.
Interpretation:
The expected value is not necessarily an outcome that will definitely occur in a single trial, but rather what you would expect to see on average over many repetitions of the experiment.
Example:
Rolling a dice: If you roll a fair six-sided die, the expected value is 3.5 because each number from 1 to 6 has a probability of 1/6, so the calculation is (11/6) + (21/6) + ... + (6*1/6) = 3.5.
'''
```

```python
#18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np

def calculate_z_score(data):
  # Mean of the dataset
    mean = np.mean(data)
   # Standard Deviation of tha dataset
    std_dev = np.std(data)
   # Z-score of tha data points
    z_scores = (data - mean) / std_dev
    return z_scores

# Example dataset
dataset = [3,9, 23, 43,53, 4, 5,30, 35, 50, 70, 150, 6, 7, 8, 9, 10]
z_scores = calculate_z_score(dataset)
print('Z-Score :',z_scores)

# Data points which lies outside 3 standard deviatioms are outliers
# i.e outside range of99.73% values
outliers = [data_point for data_point, \
          z_score in zip(dataset, z_scores) if z_score > 3]
print(f'\nThe outliers in the dataset is {outliers}')
```

```python
#19.  Implement hypothesis testing using Z-statistics for a sample dataset
import scipy.stats as stats
import math
import numpy as np

# Given data
sample_data = np.array(
    [205, 198, 210, 190, 215, 205, 200, 192, 198, 205, 198, 202, 208, 200, 205, 198, 205, 210, 192, 205, 198, 205, 210, 192, 205])
population_std_dev = 5
population_mean = 200
sample_size = len(sample_data)

# Step 1: Define the Hypotheses
# Null Hypothesis (H0): The average cholesterol level in a population is 200 mg/dL.
# Alternate Hypothesis (H1): The average cholesterol level in a population is different from 200 mg/dL.

# Step 2: Define the Significance Level
alpha = 0.05  # Two-tailed test

# Critical values for a significance level of 0.05 (two-tailed)
critical_value_left = stats.norm.ppf(alpha/2)
critical_value_right = -critical_value_left

# Step 3: Compute the test statistic
sample_mean = sample_data.mean()
z_score = (sample_mean - population_mean) / \
    (population_std_dev / math.sqrt(sample_size))

# Step 4: Result
# Check if the absolute value of the test statistic is greater than the critical values
if abs(z_score) > max(abs(critical_value_left), abs(critical_value_right)):
    print("Reject the null hypothesis.")
    print("There is statistically significant evidence that the average cholesterol level in the population is different from 200 mg/dL.")
else:
    print("Fail to reject the null hypothesis.")
    print("There is not enough evidence to conclude that the average cholesterol level in the population is different from 200 mg/dL.")
```

```python
#27.  Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
data = np.array([1, 2, 5, 8, 10, 12, 15, 18, 20, 22])

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Z-score normalization
z_scores = (data - mean) / std_dev

# Visualization
plt.figure(figsize=(8, 5))
plt.subplot(1, 2, 1)
plt.hist(data, bins=10, edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')

plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=10, edgecolor='black')
plt.title('Z-Score Normalized Data')
plt.xlabel('Z-Score')
plt.tight_layout()
plt.show()
```

```python
#28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
# Sample dataset
data = np.array([1, 2, 5, 8, 10, 12, 15, 18, 20, 22])

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Z-score normalization
z_scores = (data - mean) / std_dev

print(f'skewness:{skew(z_scores, axis=0, bias=True)}')
print(f'kurtosis: {kurtosis(z_scores, axis=0, fisher=True, bias=True)}')
```
